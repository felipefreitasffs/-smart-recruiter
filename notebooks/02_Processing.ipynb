```python
df = df.iloc[:25000]
```

```python
cols_perfil = [
    "app_cv_pt",
    "app_cv_en",
    "app_infos_basicas_objetivo_profissional",
    "app_informacoes_profissionais_titulo_profissional",
    "app_informacoes_profissionais_area_atuacao",
    "app_informacoes_profissionais_conhecimentos_tecnicos",
    "app_informacoes_profissionais_certificacoes",
    "app_informacoes_profissionais_outras_certificacoes",
    "app_informacoes_profissionais_qualificacoes",
    "app_informacoes_profissionais_experiencias",
    "app_formacao_e_idiomas_nivel_academico",
    "app_formacao_e_idiomas_nivel_ingles",
    "app_formacao_e_idiomas_nivel_espanhol",
    "app_formacao_e_idiomas_outro_idioma",
    "app_formacao_e_idiomas_instituicao_ensino_superior",
    "app_formacao_e_idiomas_cursos",
    "app_formacao_e_idiomas_outro_curso"
]
```

```python
for col in cols_perfil:
    df[col] = df[col].fillna("")

df["perfil_candidato"] = df[cols_perfil].agg(" ".join, axis=1)
```

```python
df["perfil_candidato"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df["perfil_tecnico"] = df["app_informacoes_profissionais_conhecimentos_tecnicos"].fillna("") + " " + df["app_formacao_e_idiomas_nivel_ingles"].fillna("")

vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["perfil_tecnico"])
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopwords_pt = stopwords.words('portuguese')

vectorizer = TfidfVectorizer(stop_words=stopwords_pt)
matriz_tfidf = vectorizer.fit_transform(df["perfil_candidato"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

matriz_similaridade = cosine_similarity(matriz_tfidf)
```

```python
def recomendar_candidatos_similares(indice_base, top_n=5):
    similares = list(enumerate(matriz_similaridade[indice_base]))
    similares = sorted(similares, key=lambda x: x[1], reverse=True)
    similares = similares[1:top_n+1]  # remove o próprio candidato

    recomendados = df.iloc[[i[0] for i in similares]][["nome", "perfil_candidato"]].copy()
    recomendados["similaridade"] = [i[1] for i in similares]
    return recomendados
```

```python

recomendar_candidatos_similares(2, top_n=5)
```

```python
cols_vaga = [
    "job_informacoes_basicas_titulo_vaga",
    "job_informacoes_basicas_objetivo_vaga",
    "job_perfil_vaga_areas_atuacao",
    "job_perfil_vaga_principais_atividades",
    "job_perfil_vaga_competencia_tecnicas_e_comportamentais",
    "job_perfil_vaga_habilidades_comportamentais_necessarias"
]
```

```python
for col in cols_vaga:
    df[col] = df[col].fillna("")

df["perfil_vaga"] = df[cols_vaga].agg(" ".join, axis=1)
```

```python
df["perfil_vaga"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

stopwords_pt = stopwords.words('portuguese')

vectorizer = TfidfVectorizer(stop_words=stopwords_pt, max_features=3000)

tfidf_candidatos = vectorizer.fit_transform(df["perfil_candidato"])

tfidf_vagas = vectorizer.transform(df["perfil_vaga"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def recomendar_para_vaga(indice_vaga, top_n=5):
    vetor_vaga = tfidf_vagas[indice_vaga]
    similaridades = cosine_similarity(vetor_vaga, tfidf_candidatos).flatten()
    top_indices = np.argsort(similaridades)[::-1][:top_n]

    resultados = df.iloc[top_indices][["nome", "perfil_candidato"]].copy()
    resultados["similaridade"] = similaridades[top_indices]
    return resultados
```

```python

recomendar_para_vaga(10, top_n=5)
```

```python

df["id_vaga_interacao"] = df["id_vaga_interacao"].astype(str).fillna("")

# Selecionar colunas únicas da vaga
vagas_unicas = df.drop_duplicates(subset="id_vaga_interacao").reset_index(drop=True)
```

```python
recomendacoes_todas = []

for idx, vaga in vagas_unicas.iterrows():
    id_vaga = vaga["id_vaga_interacao"]
    vetor_vaga = tfidf_vagas[idx]

    similaridades = cosine_similarity(vetor_vaga, tfidf_candidatos).flatten()
    top_indices = np.argsort(similaridades)[::-1][:5]

    candidatos = df.iloc[top_indices][["nome", "perfil_candidato"]].copy()
    candidatos["similaridade"] = similaridades[top_indices]
    candidatos["id_vaga_interacao"] = id_vaga
    candidatos["titulo_vaga"] = vaga["job_informacoes_basicas_titulo_vaga"]

    recomendacoes_todas.append(candidatos)
```

```python
df_recomendacoes = pd.concat(recomendacoes_todas, ignore_index=True)
```
